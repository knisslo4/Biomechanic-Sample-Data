```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
Pitch_1 = pd.read_csv('/Users/logankniss/Desktop/PitcherY_Pitch1.csv')
Pitch_1.head()
```

```python
Pitch_1.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler

N_Pitch_1 = Pitch_1.copy()
  
column = 'time'
N_Pitch_1[column] = MinMaxScaler().fit_transform(np.array(N_Pitch_1[column]).reshape(-1,1))
   
display(N_Pitch_1)
```

```python
from sklearn.ensemble import IsolationForest

column_data = N_Pitch_1[['right_shoulder_external_rotation']].values
clf = IsolationForest(contamination=0.05)  # contamination determines the proportion of outliers in the dataset
predictions = clf.fit_predict(N_Pitch_1[['right_shoulder_external_rotation']])
predictions
df_cleaned = N_Pitch_1[predictions != -1]
df_cleaned.head()
```

```python
plt.plot(df_cleaned['time'], df_cleaned['right_shoulder_external_rotation'])
```
